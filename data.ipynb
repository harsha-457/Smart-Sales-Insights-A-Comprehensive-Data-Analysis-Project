```python
#T TEST
price_a = products.loc[products['Product Name'].str.contains('MP3', na=False), 'Unit Price USD']
price_b = products.loc[products['Product Name'].str.contains('Player', na=False), 'Unit Price USD']

t_stat, p_value = stats.ttest_ind(price_a, price_b, nan_policy='omit')

print(f"T-test results:\nT-statistic: {t_stat:.4f}\nP-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("\nConclusion: The difference in means is statistically significant (reject null hypothesis).")
else:
    print("\nConclusion: No statistically significant difference in means (fail to reject null hypothesis).")
```

```python
insights = {
    't_statistic': round(t_stat, 4),
    'p_value': round(p_value, 4),
    'mean_price_a': round(price_a.mean(), 2),
    'mean_price_b': round(price_b.mean(), 2),
    'std_dev_price_a': round(price_a.std(), 2),
    'std_dev_price_b': round(price_b.std(), 2),
    'sample_size_price_a': len(price_a),
    'sample_size_price_b': len(price_b)
}
print("\n===== Statistical Insights =====")
print(f"T-statistic: {insights['t_statistic']}")
print(f"P-value: {insights['p_value']}")
print(f"Mean Price (Group A - MP3): {insights['mean_price_a']} USD")
print(f"Mean Price (Group B - Player): {insights['mean_price_b']} USD")
print(f"Standard Deviation (Group A - MP3): {insights['std_dev_price_a']} USD")
print(f"Standard Deviation (Group B - Player): {insights['std_dev_price_b']} USD")
print(f"Sample Size (Group A - MP3): {insights['sample_size_price_a']}")
print(f"Sample Size (Group B - Player): {insights['sample_size_price_b']}")


alpha = 0.05
if insights['p_value'] < alpha:
    print("\nConclusion: The difference in means is statistically significant (reject null hypothesis).")
else:
    print("\nConclusion: No statistically significant difference in means (fail to reject null hypothesis).")
```

```python
# Calculate average quantity sold by gender
avg_quantity_by_gender = merged_data.groupby('Gender')['Quantity'].mean()

# Calculate additional statistics
total_quantity_by_gender = merged_data.groupby('Gender')['Quantity'].sum()
count_by_gender = merged_data['Gender'].value_counts()

# Perform a T-test between genders
male_sales = merged_data[merged_data['Gender'] == 'Male']['Quantity']
female_sales = merged_data[merged_data['Gender'] == 'Female']['Quantity']
t_stat, p_value = stats.ttest_ind(male_sales, female_sales, nan_policy='omit')

# Prepare insights
insights = {
    'average_quantity_by_gender': avg_quantity_by_gender.round(2).to_dict(),
    'total_quantity_by_gender': total_quantity_by_gender.to_dict(),
    'sample_size_by_gender': count_by_gender.to_dict(),
    't_statistic': round(t_stat, 4),
    'p_value': round(p_value, 4),
}

# Display insights
print("\n===== Statistical Insights =====")
print(f"Average Quantity Sold by Gender: {insights['average_quantity_by_gender']}")
print(f"Total Quantity Sold by Gender: {insights['total_quantity_by_gender']}")
print(f"Sample Size by Gender: {insights['sample_size_by_gender']}")
print(f"T-Statistic: {insights['t_statistic']}")
print(f"P-Value: {insights['p_value']}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: The difference in quantity sold between genders is statistically significant (reject null hypothesis).")
else:
    print("\nConclusion: No statistically significant difference in quantity sold between genders (fail to reject null hypothesis).")
```

```python
# Filter data for Key Item values between 0 and 1000
filtered_data = merged_data[(merged_data['StoreKey'] >= 0) & (merged_data['StoreKey'] <= 1000)]

# Calculate Converted Sales (standardized currency values)
if 'Exchange Rate' in filtered_data.columns:
    filtered_data['Converted Sales'] = filtered_data['Order Value'] * filtered_data['Exchange Rate']
else:
    filtered_data['Converted Sales'] = filtered_data['Order Value']  # Assume already in base currency

# Group by Currency Code and calculate total sales (both original and converted)
sales_by_currency = filtered_data.groupby('Currency Code').agg({
    'Order Value': 'sum', 
    'Converted Sales': 'sum'
}).reset_index()

# Sort data for visualization
sales_by_currency_sorted = sales_by_currency.sort_values(by='Converted Sales', ascending=False)

# Plot: Total Converted Sales by Currency
plt.figure(figsize=(12, 6))
sns.barplot(
    data=sales_by_currency_sorted, 
    x='Currency Code', 
    y='Converted Sales', 
    palette='coolwarm'
)
plt.title('Total Sales by Currency (Key Item 0-1000)', fontsize=16)
plt.xlabel('Currency Code', fontsize=12)
plt.ylabel('Total Converted Sales (Base Currency)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot: Total Original Sales by Currency
plt.figure(figsize=(12, 6))
sns.barplot(
    data=sales_by_currency_sorted, 
    x='Currency Code', 
    y='Order Value', 
    palette='viridis'
)
plt.title('Total Sales by Currency (Original, Key Item 0-1000)', fontsize=16)
plt.xlabel('Currency Code', fontsize=12)
plt.ylabel('Total Sales (Original Currency)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
